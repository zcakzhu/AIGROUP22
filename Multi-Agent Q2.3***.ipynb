```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from random import randint

# Displaying all columns and rows 
pd.options.display.max_columns = 10000
pd.options.display.max_rows = 10000000
```

```python
valid= pd.read_csv('validation.csv')
```

```python
n=50
multi = pd.DataFrame(columns = ['Imps','Click','Spend','CTR','CPM','eCPC'])

spend =[0 for i in range (n)]
imps = [0 for i in range (n)]
click = [0 for i in range (n)]
iteration = 0   

bidprice = np.zeros((len(valid),50))
for i in range(len(valid)):
    for j in range(50):
        bidprice[i,j] = np.random.randint(160000, 180000)/1000

bidprice_pd = pd.DataFrame(bidprice)

a = bidprice_pd.apply(lambda row: row.nlargest(2).values[-1],axis=1) #找到第二大数
aa = a.values
b = valid.payprice.values # 把pd 转成 matrix 方便下面一步的计算
c = np.maximum(a,b) # 找element wise 最大的数，shape = number of impressions * 1

for i in range(n):     # iterate through 50个 bidding agents
    iteration +=1
    print('strategy:',iteration)
    for j in range(len(valid)):  
        # iterate through 每一个impression
# #             if budget[i] < valid.payprice[j]/1000:
# #                 continue
# #             if budget[i] < a[j]/1000:
# #                 continue
        if bidprice_pd.iloc[j,i] >= valid.payprice[j] and bidprice_pd.iloc[j,i] > aa[j]:
            spend[i] = spend[i] + c[j]
#             imps[i] += 1
#             if valid.click[j]==1:
#                 click[i] += 1
            if spend[i]>6250000:
                spend[i] = spend[i] - c[j]

            click[i] = click[i]+valid.click[j]
            imps[i] = imps[i]+1

#feeding data into a dataframe
#multi.loc[iteration,'Base_Bid'] = bidprice[i]
    multi.loc[iteration,'Imps'] = imps[i]
    multi.loc[iteration,'Click'] = click[i]
    multi.loc[iteration,'Spend'] = spend[i]/1000

multi['CTR'] = multi['Click']/multi['Imps']
multi['eCPC'] = multi['Spend']/multi['Click']
multi['CPM'] =  multi['Spend']*1000/multi['Imps']
multi.loc['Average',:]= multi.mean(axis=0)
```

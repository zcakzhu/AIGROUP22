```python
multi = pd.DataFrame(columns = ['Imps','Click','Spend','CTR','CPM','eCPC'])

n=12
    
win = np.zeros((len(ALL),n))

#flag = True
spend = [0 for i in range (n)]
imps = [0 for i in range(n)]
click = [0 for i in range(n)]
iteration =0  
#iteration1 =0 
    

for j in range(n):
    iteration +=1 
    print('Running Strategy' + str(iteration))
    for i in range(len(ALL)):
        #iteration1 +=1
        #print('Running Strategy' + str(iteration) + ':' + str(iteration1) + '/' + str(len(ALL)*n))
        
        #if budget[j] < condition_min[i]/1000:
#         if budget[j] < ALL.payprice[i]/1000:
#             continue
#         if budget[j] < secondlargestbidprice[i]/1000:
#             continue
        if ALL_drop.iloc[i,j] > secondlargestbidprice[i] and ALL_drop.iloc[i,j] >= ALL.payprice[i]:
            spend[j] = spend[j] + secondlargestbidprice_payprice[i]
                
                    #imps[j] += 1
                    #if ALL.click[i] == 1:
                        #click[j] += 1
            if spend[j]>6250000:
                spend[j] = spend[j] - secondlargestbidprice_payprice[i]
                    #flag = False
                    #break
                    
            click[j] = click[j] + ALL.click[i]
            imps[j] = imps[j] + 1
            win[i,j] = 1
                
    #feeding data into a dataframe
    multi.loc[iteration,'Imps'] = imps[j]
    multi.loc[iteration,'Click'] = click[j]
    multi.loc[iteration,'Spend'] = spend[j]/1000

# multi['CTR'] = multi['Click']/multi['Imps']
# multi['eCPC'] = multi['Spend']*1000/multi['Click']
# multi['CPM'] =  multi['Spend']/multi['Imps']

multi1 = multi
multi1['Strategy'] = ['LRLINEAR','LRQUAD','LRORTB','LREXP','GBCLINEAR','GBCQUAD',
                      'GBCORTB','GBCEXP','GBCCUBE','GBCFOUR','GBCFIF','GBCKAN']
                      
multi1.head()
```
